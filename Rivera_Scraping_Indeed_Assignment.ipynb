```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://www.indeed.com/jobs?q=front+end+developer&l=Florida&from=searchOnHP&vjk=9daa6da0b5835660');

page = driver.page_source

soup = BeautifulSoup(page,'html.parser')
```

```python
#id_list = []
jk_list = []
link_list = []
job_list = []
company_list = []
location_list = []
#salary_list = []

#get the job ids (15 per page)
tables = soup.find_all("td", class_="resultContent")



def find_job_title():
    job_a = table.find('a')
    job_title = job_a.find('span').text
    job_list.append(job_title)
    
def find_company_name():
    company_name = table.find('span', class_='companyName').text
    company_list.append(company_name)

def find_location():
    company_location = table.find('div', class_='companyLocation').text
    location_list.append(company_location)
    
def find_job_link():
    job_a = table.find('a')
    job_jk = job_a['data-jk']
    link_list.append("https://www.indeed.com/viewjob?jk=" + job_jk + "&tk=1gr98572ejiie801&from=serp&vjs=3")
    
        
for table in tables:
    find_job_title()
    find_company_name()
    find_location()
    find_job_link()
    print ('jobs:' + str(len(job_list))+'company:' + str(len(company_list))+'location:' + str(len(location_list))+'Links:' + str(len(link_list)))

#these work:
#print(job_list)
#print(link_list)
#print(company_list)
#print(location_list)



#function to get the ids from one page
#run it
```
